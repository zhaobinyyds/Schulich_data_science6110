```python
apple = electronics_purchases[electronics_purchases['brand'] == 'apple']['price']
all_other = electronics_purchases[electronics_purchases['brand'] != 'apple']['price']

t_stat, p_val = stats.ttest_ind(apple, all_other)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}")
```
