```python
import pandas as pd
```

```python
#read the file from the local computer
df = pd.read_csv('/Users/angelis/Downloads/sales_data.csv')
```

```python
# check the data frame 
df
```

```python
#check the information of the dataframe 
df.info()
```

```python
#check the number and number of null value of data for each column and 
df.describe(include='all')
```

```python
#data cleaning started
#find the null value in Price and Units Sold column and fill them in with the average value of these two column. Because the Sale Date column has no null value, so there is no need to fill.
df['Price']=df['Price'].fillna(df['Price'].mean())
df['Units Sold']=df['Units Sold'].fillna(df['Units Sold'].mean())
```

```python
# after filling the null value in these two column, check the number of data in Price and Units Sold column is 1050 or not.
df.describe(include='all')
```

```python
# drop the row with the null value in Product and Region column, and describe the dataframe after that.
df = df.dropna()
df.describe(include = 'all')
```

```python
#check if there is no null remain in the dataframe
df.info()
```

```python
# cleaning the dataframe by dropping the duplications 
df = df.drop_duplicates()
df
```

```python
# Adding a new column Revenue which is the product of Price and Units sold 
df['Revenue']= df['Price'] * df['Units Sold']

df
#finished Data cleaning 
```

```python
#Check the dataframe after remove the duplication and adding the new column
df.describe()
```

```python
# Finding the maximum value of Revenue in the dataframe 
df.loc[df['Revenue']==df['Revenue'].max()]
```

```python
#Group the dataframe by Region 
df_grouped = df.groupby('Region')
```

```python
#Check the Maximum and Minimum value of Price, Units Sold and Revenue for each region 
df_grouped.describe()
```
